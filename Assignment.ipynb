**Question 1: What is Data Science?**
 - Ans: Data Science is the process of collecting and analyzing data to find useful patterns and meaningful insights. It helps people make better decisions using data instead of assumption.

** Why is it more than just statistics or programming? **

- Ans: Data Science is more than just statistics or programming because it combines both with domain knowledge to solve real-world problems. It’s about understanding the problem, working with data, and communicating insights—not just writing code or doing calculations.

- For example:, A company like Amazon, Flipkart, Netflix uses Data Science to recommend products to customers. It combines data analysis (statistics), coding (programming), and understanding customer’s behavior to suggest what they might want to buy next.

```python
#Question 10: A startup collects customer feedback as text and numerical ratings. Explain step-by-step:
#1.	How raw text becomes structured data.?

feedback = [
    "Great product, very fast delivery!",
    "Bad quality, not worth the price",
    "Average experience, okay service"
]

ratings = [5, 2, 3]

import re

cleaned_feedback = []

for text in feedback:
    text = text.lower()  # convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    cleaned_feedback.append(text)

print(cleaned_feedback)

tokenized = [text.split() for text in cleaned_feedback]
print(tokenized)

stopwords = ["the", "is", "and", "very", "not"]

filtered = []
for words in tokenized:
    filtered.append([w for w in words if w not in stopwords])

print(filtered)

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform([' '.join(words) for words in filtered])

print(X.toarray())
print(vectorizer.get_feature_names_out())

import pandas as pd

df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df["rating"] = ratings

print(df)


```
